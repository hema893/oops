```python
try:
    a = 10
    b = 0
    print(a / b)
except ZeroDivisionError:
    print("Division by zero is not allowed")
```

```python
try:
    a = 10
    b = 2
    print(a / b)
except ZeroDivisionError:
    print("Error")
else:
    print("Division successful")
finally:
    print("This will always execute")
```

```python
 5✅ Custom Exception:

User-defined exception class ko custom exception kehte hain.

✅ Why needed?

Specific business logic errors handle karne ke liye

Code ko clean aur meaningful banane ke liye
```

```python
class NegativeNumberError(Exception):
    pass

num = -10

if num < 0:
    raise NegativeNumberError("Negative number not allowed")
```
