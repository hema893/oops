```python
1 ✅ Abstraction:

Abstraction ka matlab hai implementation details ko hide karna aur sirf essential features user ko dikhana.

Real-life example:
Car chalane ke liye aapko engine ka internal working nahi pata hona chahiye — sirf steering, brake, accelerator use karte ho.
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius * self.radius

c1 = Circle(5)
print(c1.area())
```

2 | Abstraction                            | Encapsulation                                   |
| -------------------------------------- | ----------------------------------------------- |
| Implementation hide karta hai          | Data ko protect karta hai                       |
| Abstract class/interface use karta hai | Private variables & getter-setter use karta hai |
| Focus: What to show                    | Focus: How to protect                           |


```python
class Student:
    
    def __init__(self, name):
        self.__name = name   # private variable
    
    def get_name(self):
        return self.__name

s1 = Student("Rahul")
print(s1.get_name())
```
